```python
irisdata = pd.read_csv('C:\\Users\\GreenGreen\\Documents\\python3\\python_dataviz\\iris.data' ,header =None)
irisdata
```

### numpy จัดการตัวเลขโดยเฉพาะ

```python
setosa_np= irisdata.iloc[:50,:-1].to_numpy()
setosa_np[:5,:]
```

#### plot ค่า x กับ  y  แบบง่าย

```python
# มาร์คเกอร์ คือสัญลักษณ์ก็คือจุด 
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1])
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=6) #s ไซต์ลดขนาด
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=60)
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=setosa_np[:,2])
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2])#เพิ่มขนาดให้ทุกตัวเป็น 10 เท่า
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2],c ='g') #เปลี่ยนสี cฟ้า yเหลือง  kดำ  mม่วง  gเขียว  rแดง  bน้ำเงิน
```

```python
print(f'max = {max(setosa_np[:,3])}, min ={min(setosa_np[:,3])}')
```

```python
colors = []
for i in setosa_np[:,3]:
    if i < 0.15:
        colors.append('b')
    elif i < 0.3:
          colors.append('g')
    elif i < 0.45:
          colors.append('y')
    else:
        colors.append('r')
colors
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c =setosa_np[:,3])
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c =setosa_np[:,3],cmap = 'YlOrBr')
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c= np.array([[1,0.5,0.8]]))
```

## Visualize 2 Data

```python
versicolor_np =irisdata.iloc[50:100,:-1].to_numpy()
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c =setosa_np[:,3],cmap = 'BuPu',marker='p')
plt.scatter(versicolor_np[:,0],versicolor_np[:,1],s=30*versicolor_np[:,2],c =versicolor_np[:,3],cmap = 'BuPu',marker='*')
```

```python
virginia_np=irisdata.iloc[:100,:-1].to_numpy()
```

## visualize 3 data

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2],c =setosa_np[:,3],cmap = 'BuPu',marker='p')
plt.scatter(versicolor_np[:,0],versicolor_np[:,1],s=30*versicolor_np[:,2],c =versicolor_np[:,3],cmap = 'BuPu',marker='*')
plt.scatter(virginica_np[:,0],virginica_np[:,1],s=30*virginica_np[:,2],c =virginica_np[:,3],cmap = 'BuPu',marker='x')
```
