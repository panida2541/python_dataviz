```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df =pd.read_excel('C:\\Users\\GreenGreen\\Documents\\python3\\python_dataviz\\income.xls',header=3) #กำหนดหัวตารางเอาตารางว่างออก
df
```

```python
#ไอร็อคตัดตาราง
greaterbargkok = df.iloc[4:8,:]  #ใส่เลขเกิน 1 เลข
greaterbargkok
```

## Barchart(กราฟแท่ง)

## สร้างกราฟแท่งเปรียบเทียบรายได้ในปี 2541 ของจังหวัดกรุงเทพและปริมณฑล

```python
plt.bar([1,2,3,4],greaterbargkok.iloc[:,2])#เลือกทุกแถว เลือกคอลัมน์ที่2
plt.title(u'กราฟแท่งเปรียบเทียบรายได้ในปี 2541 ของจังหวัดกรุงเทพและปริมณฑล',fontname='Tahoma',fontsize='18')  #ตั้งชื่อกราฟ
plt.xlabel('Provinces')
plt.ylabel('Income Eath/Month/Household')
```

```python
plt.bar([1,2,3,4],greaterbargkok.iloc[:,2],tick_label=greaterbargkok.iloc[:,-1])#เลือกทุกแถว เลือกคอลัมน์ที่2
plt.title('Barchart comparing average monthly income of provinces in Greater Bangkok in 2541')  #ตั้งชื่อกราฟ
plt.xlabel('Provinces')
plt.ylabel('Income Eath/Month/Household')
```

## compare different years

```python
plt.bar([1-(0.35/2),2-(0.35/2),3-(0.35/2),4-(0.35/2)],greaterbargkok.iloc[:,-2],width=0.35,tick_label=greaterbargkok.iloc[:,-1],color = 'r',label ='2558',align='edge')
plt.bar([1+(0.35/2),2+(0.35/2),3+(0.35/2),4+(0.35/2)],greaterbargkok.iloc[:,2],width=0.35,tick_label=greaterbargkok.iloc[:,-1],color = 'b',label ='2541',align='edge')
plt.title('Barchart comparing average monthly income of provinces in Greater Bangkok in 2541,2558')  #ตั้งชื่อกราฟ
plt.xlabel('Provinces')
plt.legend(loc=1)#บอกปีและตำแหน่งการแสดง
plt.ylabel('Income Eath/Month/Household')
```
